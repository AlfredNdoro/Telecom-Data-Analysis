```python
#Loading the cells_geo Dataset
df = pd.read_csv('cells_geo.csv', delimiter=',')

#Splitting the columns into their respective ones as the data only had one column
df[';VILLES;STATUS;LOCALISATION;DECOUPZONE;ZONENAME;LONGITUDE;LATITUDE;REGION;AREA;CELL_ID;SITE_CODE'].str.split(';')

#Asigning the name SPLIT to it
SPLIT = df[';VILLES;STATUS;LOCALISATION;DECOUPZONE;ZONENAME;LONGITUDE;LATITUDE;REGION;AREA;CELL_ID;SITE_CODE'].str.split (';',expand = True)
df = SPLIT

#Reassigning the df dataframe name
df = SPLIT.rename(columns ={0:'Number',1:'VILLES',2:'STATUS',3:'LOCALISATION',4:'DECOUPZONE',5:'ZONENAME',6:'LONGITUDE',7:'LATITUDE',8:'REGION',9:'AREA',10:'CELL_ID',11:'SITE_CODE'})
df

#DAY I busiest cities based on SITE_ID of the cities with the highest value, taking the first 4 rows

value_list = ['f939f3349c', '5324bc6c24','47yB6si0R9','5324bc6c24']

df[df.SITE_CODE.isin(value_list)]

#DAY 2 busiest cities based on SITE_ID of the cities with the highest value, taking the first 5 rows

value_list = ['2e828e607c', 'ea868b8a04','969b8doe39','EpCRhKGWMS','f7f56b7293']

df[df.SITE_CODE.isin(value_list)]

#DAY 3 busiest cities based on SITE_ID of the cities with the highest value, taking the first 5 rows

value_list = ['a64fc672f1', 'MpOhP3OKNN','2dc799193C','ae6cdb80c1']

df[df.SITE_CODE.isin(value_list)]
```
